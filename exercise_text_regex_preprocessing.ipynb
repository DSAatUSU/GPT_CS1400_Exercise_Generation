````python
import json
import os
import re
# (\"code_snippet_[1-9]\":\s*)?(\"\\n\\n\"?\s*?[`]*?python?\\n\")([\s\S]*?)(```\\n\")|(\"code_snippet_[1-9]\":\s*)?(\"\\n\\n\")([\s\S]*?)(\"```\\n\",)
# (\"[0-9A-Z]\":\s*\")([\s\S]*?)(\s*?\"options_3\")(?=\s*\"[0-9A-Z]\":\s*\"|$)
# (\"[0-9A-Z]\":\s*\")([\s\S]*?)(?=\s*\"[0-9A-Z]\":\s*\")

# group 2 WITH r'(\"questions\":\s{\s*)([\s\S]*?)(}\s*},\s*\"solution\":)'  finds all everything within questions: {}

def preprocess_text(text):
    # Remove specific sequences and escape double quotes within values
    text = text.replace(r"```json", "")
    text = re.sub(r'\}\n```$', r'}', text)
    text = re.sub(r'\n```",', r'",', text)
    text = text.replace("```python\n", "")
    text = text.replace("```python\\n", "")
    text = re.sub(r'\\n```",', r'",', text)

    # ("lines\sof\scode":\s{\n)(\s*(\"[A-Z]\:\s).*",?\n)

    # This regex looks for all the options A: B: C: in lines of code 
    pattern = r'("lines\sof\scode":\s{\n)((?:.*\n)*?)(\s*},\n\s*"solution")'
    # Function to replace "A: " with "A": " in the second group
    def replace_label(match):
        g2 = match.group(2).replace("\n```", "")
        g2 = re.sub(r'(\"([A-Z])\:\s)+', r'"\2": "', g2)
        return match.group(1) + g2 + match.group(3)
    text = re.sub(pattern, replace_label, text)
    text = text.replace(r'"": ""', '": "')
    text = re.sub(r'(,)(\n*\s*},\n*\s*"solution")', r'\2', text)
    pattern = r'(\"provided code\":\s*\")([\s\S]*?)(\",\n\s*\"Instructions\")'
    def replace_newlines(match):
        g2 = match.group(2).replace("\n", "\\n")
        return match.group(1) + g2 + match.group(3)
    text = re.sub(pattern, replace_newlines, text)

    pattern = r'(\"solution\":\s*\")([\s\S]*?)(\",\n\s*\"prerequisite\")'
    def replace_newlines(match):
        g2 = match.group(2).replace("\n", "\\n")
        return match.group(1) + g2 + match.group(3)
    text = re.sub(pattern, replace_newlines, text)
    return text

def preprocess_error_text(text):
    text = re.sub(r'\\n```":\s\"\"|\\n```":\s{}', r'"', text)
    pattern = r'\"(sample input|sample output|task|[A-Z]|[0-9]|solution|prerequisite|prerequisites|hint|scenario|provided code|Instructions)\":\s(("?).*\n(?=\s*?}?,?\s*?\"?(sample input|sample output|task|[A-Z]|[0-9]|solution|prerequisite|prerequisites|hint|scenario|provided code|Instructions)?\"?))'
    def escape_double_quotes(match):
        def escape_double_quotes_in_values(match):
            g2 = match.group(2).replace(r'"', '\'').replace('\\', '\\\\') #MAKE THIS MORE SPECIFIC TO ALL SINGE ESCAPE CHARACTER WHICH ARE CAUSING PROBLEMS DONT JUST REPLACE EVERY ESCAPE CHARACTER
            g4 = match.groups(0)[-1]
            return f'"{g2}"{g4}'
        correct_value = re.sub(r'(")(.*?)(")(,?$\]?)', escape_double_quotes_in_values, match.groups(0)[1])
        return f'"{match.groups(0)[0]}": {correct_value}'
    text = re.sub(pattern, escape_double_quotes, text)

    pattern = r'(\"code_snippet_[1-9]\":\s*)?(\+\s*)?(\"\\n\\n\")([\s\S]*?)(\"```\\n\",)'
    def formatting_code(match):
        g1 = match.group(1)
        g4 = match.group(4).replace('\n',r'\n').replace("\"", "\'")
        if g1 is not None:
            return f'{g1}"```{g4}```",'
        else:
            return f',\n"code_snippet": "```{g4}```",'
        #return f'{match.group(1)}{correct_value}{match.group(3)}'
    text = re.sub(pattern, formatting_code, text)

    pattern = r'(\"Instructions\"[\s\S]*?)(\s*},\n)?(,)?(\s*\"solution\":)|(\"sample\soutput\"[\s\S]*?)(\s*},\n)?(,)?(\s*\"solution\":)'
    def fixing_bracket_imbalance(match):
        g1 = match.group(1)
        g2 = match.group(2)
        g4 = match.group(4)
        g5 = match.group(5)
        g6 = match.group(6)
        g8 = match.group(8)
        gx = "\n  },"
        if g5 is None:
            if g2 is None:
                return f'{g1}{gx}{g4}'
            else:
                return f'{g1}{g2}{g4}'
        if g1 is None:
            if g6 is None:
                return f'{g5}{gx}{g8}'
            else:
                return f'{g5}{g6}{g8}'
    text = re.sub(pattern, fixing_bracket_imbalance, text)
    return text

def preprocess_escape_characters(text):
    def replace_backslashes(match):
        char = match.group(1)
        return char if char in ['/',"\b","\f","\t","\r","\n"] else '\\\\'

    # Use re.sub() with a regex pattern to replace backslashes
    text = re.sub(r'\\(.)', replace_backslashes, text)
    return text

def preprocess_invalid_control_characters(text):
    text = re.sub(r'\\",', r'",', text)
    pattern = r'(}?,\s*\"solution\":\s\")([\s\S]*?)(\")(\s*?}$)(\s*```)?'
    def replace_invalid_control_characters(match):
        g1 = match.group(1)
        g2 = match.group(2).replace('\n',r'\n').replace("\"", "\'")
        g3 = match.group(3)
        g4 = match.group(4)
        return f'{g1}```{g2}```"{g4}'
    text = re.sub(pattern, replace_invalid_control_characters, text)
    pattern = r'(\"Instructions\"[\s\S]*?)(\s*},\n)?(,)?(\s*\"solution\":)|(\"sample\soutput\"[\s\S]*?)(\s*},\n)?(,)?(\s*\"solution\":)'
    def fixing_bracket_imbalance(match):
        g1 = match.group(1)
        g2 = match.group(2)
        g4 = match.group(4)
        g5 = match.group(5)
        g6 = match.group(6)
        g8 = match.group(8)
        gx = "\n  },"
        if g5 is None:
            if g2 is None:
                return f'{g1}{gx}{g4}'
            else:
                return f'{g1}{g2}{g4}'
        if g1 is None:
            if g6 is None:
                return f'{g5}{gx}{g8}'
            else:
                return f'{g5}{g6}{g8}'
    text = re.sub(pattern, fixing_bracket_imbalance, text)
    return text

def preprocess_extra_data(text):
    pattern = r'(})(\s*)?(```)(\s*?)$'
    def remove_extra_data(match):
        return match.group(1)
    text = re.sub(pattern, remove_extra_data, text)
    return text


def read_json_file(file_path, counter, error_counter):

    if not os.path.exists(file_path):
        print(f"File not found: {file_path}")
        return

    if os.path.getsize(file_path) == 0:
        print(f"File is empty: {file_path}")
        return

    try:
        with open(file_path, 'r') as file:
            file_content = file.read()
            preprocessed_content = preprocess_text(file_content)
            #data = loads(preprocessed_content, ALL)
            data = json.loads(preprocessed_content)
            #data = process_json_like_content(preprocessed_content)
            return data
    except json.JSONDecodeError as e:
        if "Expecting ',' delimiter" in str(e):
            print(f"{error_counter},Expecting ',' delimiter")
            further_preprocessed_content = preprocess_error_text(preprocessed_content)
            try:
                data = json.loads(further_preprocessed_content)
                error_counter += 1
                return data
            except json.JSONDecodeError as sub_e:
                 print(f"{counter} | FILE ORIGINAL CONTENT: {file_content}\n")
                 print(f"ERROR IN FURTHER PROCESSING", sub_e)
                 print(further_preprocessed_content)
        if "Invalid \\escape" in str(e):
            print(f"{error_counter},Invalid \\escape")
            further_preprocessed_content = preprocess_escape_characters(preprocessed_content)
            try:
                data = json.loads(further_preprocessed_content)
                return data
            except json.JSONDecodeError as sub_e:
                print(f"*" * 100)
                print(f"{counter} | FILE ORIGINAL CONTENT: {file_content}\n")
                print(f"ERROR IN FURTHER PROCESSING", sub_e)
                char_num = int(re.findall(r'\d+', str(sub_e).upper())[-1])
                print(f"ERROR AT: {preprocessed_content[char_num - 50: char_num + 50]}\n\n")
                print(further_preprocessed_content)
        if "Invalid control character" in str(e):
            print(f"Invalid control character")
            further_preprocessed_content = preprocess_invalid_control_characters(preprocessed_content)
            try:
                data = json.loads(further_preprocessed_content)
                return data
            except json.JSONDecodeError as sub_e:
                print(f"{counter} | FILE ORIGINAL CONTENT: {file_content}\n")
                print(f"ERROR IN FURTHER PROCESSING", sub_e)
                print(further_preprocessed_content)
        if "Extra data" in str(e):
            print(f"Extra data")
            further_preprocessed_content = preprocess_extra_data(preprocessed_content)
            try:
                data = json.loads(further_preprocessed_content)
                return data
            except json.JSONDecodeError as sub_e:
                print(f"{counter} | FILE ORIGINAL CONTENT: {file_content}\n")
                print(f"ERROR IN FURTHER PROCESSING", sub_e)
                print(preprocessed_content)
        with open("json_parse_error_files_logs.txt", "a") as file:    
            file.write(f"{counter} | FILE ORIGINAL CONTENT: {file_content}\n")
            file.write(f"FILE PATH: {file_path[19:].split('/')}\n")
            file.write(f"ERROR DECODING JSON: {str(e).upper()}\n")
            char_num = int(re.findall(r'\d+', str(e).upper())[-1])
            file.write(f"{counter} | PREPROCESSED TEXT: {preprocessed_content}\n\n")
            file.write(f"ERROR AT: {preprocessed_content[char_num - 50: char_num + 50]}\n\n")
            file.write("*^*" * 100 + "\n\n")
    except Exception as e:
        with open("json_parse_error_files_logs.txt", "a") as file:
            file.write(f"{counter} | FILE ORIGINAL CONTENT: {file_content}\n")
            file.write(f"AN ERROR OCCURED: {str(e).upper()}\n")
            file.write(f"{counter} | PREPROCESSED TEXT: {preprocessed_content}\n\n")
            file.write("*^*" * 100 + "\n\n")

def read_files_from_df(df):
    """Reads the txt files from the dataframe and returns a list of the files"""
    list_of_files = {}
    counter = 0
    error_counter = 0
    with open("json_parse_error_files_logs.txt", "w") as file:
        file.write("")
    for i, row in df.iterrows():
        dir = row["main_dir"] + "/" + row["subdirectory"] + "/" + row["category"] + "/" + row["subcategory"] + "/" + row["excercise_type"]+ "/"+ row["difficulty"] + "/" + row["filename"]
        list_of_files[row["excercise_type"] + "`" + row["category"] + "`" + row['subcategory'] + "`" + row['difficulty'] + "`" + row['filename']] = read_json_file(dir, counter, error_counter)
        counter += 1
    return list_of_files
display(completed_files_df.iloc[:])
temp_files = read_files_from_df(completed_files_df.iloc[:])
display(temp_files)
````

````python
"{match.group(2).replace('\\"', "\\'")}"
# Escape single quotes and replace double quotes with single quotes
# value = re.sub(r"(?<!\\)'", "'", value)
def preprocess_text(text):
    # Remove specific sequences and escape double quotes within values
    text = text.replace("```json", "")#.replace("```", "")#.replace('\\"', "'")
    
    text = text.replace(": \"\"", "")
    text = text.replace(": {}", "")
    #text = re.sub(r"\"([A-Z1-9])": : \"([A-Z1-9])\"","", text)

    
    # Replacement function
    def replace_double_quote_with_single_mcq(match):
    text = re.sub(r'(\"lines\sof\scode\"\:\s{\n.*)(\"A\:\s)(.*\n.*)(.*\"B\:\s)(.*\n.*)(.*\"C\:\s)(.*\n.*)(.*\"D\:\s)', replace_double_quote_with_single_mcq, text)

    # Replacement \" with \'
    def replace_double_quote_with_single_coding(match):
        g1 = f"{match.group(1)}"
        g2 = match.group(2).replace('\\"', "\\\\'")
        return f'"{g1}": "{g2}"'
    text = re.sub(r'"([A-Z1-9])": "(.*?)(?<!\\)"', replace_double_quote_with_single_coding, text)

    # Replace new lines in values with \n and escape single quotes
    def replace_newlines_and_escape_quotes(match):
        value = match.group(1).replace('"', "'").replace('\n', '\\n')
        return f'"{value}"'
    text = re.sub(r'"\s*([^"]+)\s*"', replace_newlines_and_escape_quotes, text)

    # Correct the format of 'lines of code'
    def correct_lines_of_code(match):
        key = match.group(1)
        value = match.group(2)
        return f'"{key}": "{value}"'

    text = re.sub(r'"([A-Z1-9]): ([^"]+)"', correct_lines_of_code, text)

    
    text = text.replace("```python\\n", "").replace("```", "")

    def replace_function(match):
        key = match.group(1)
        value = match.group(2)
        return f'"{key}": "{value}",'

    text = re.sub(r'"\s*([A-Z]):\s*(.*?)\\n"\s*,?', replace_function, text)
    
    def drop_endpart_of_choices(match):
        # Drop the last useless part of the match e.g : "A",
        value = f"\"{match.group(2)}\": \"{match.group(3)}\""
        return value
    text = re.sub(r'("([A-Z])": ")([^"]*)(",:\s*"[A-Z]")', drop_endpart_of_choices, text)
    # Drop the "," at the end of the last choice

    def replace_escape_sequences(match):
        key = match.group(1)
        value = match.group(2).replace(r'^\\', "")          #.replace('\\(', '\\\\(').replace('\\)', '\\\\)').replace('"', "'")
        return f'{key}{value}"'
    text = re.sub(r'(".*?":\s*")([^"]*)"', replace_escape_sequences, text)

    

    # Regular expression pattern and replacement
    pattern = r'("[0-9]":\s*")([^"]+)"\s+or\s+"([^"]+)(")'
    replacement = r'\1\2 or \3\4'

    # Perform the substitution
    text = re.sub(pattern, replacement, text)
    text = re.sub(r',\s*}', '}', text)

    return text
````

````python
# Function definitions
def process_json_like_content(text):
    # Split the text into lines
    lines = text.splitlines()

    # Initialize variables
    key = None
    value = []
    data = {}

    # Iterate through each line
    for line in lines:
        # Check if the line contains a key
        if line.strip().endswith(': {') or line.strip().endswith(': "'):
            # Save the previous key-value pair if exists
            if key is not None:
                data[key] = '\n'.join(value).strip(' "')

            # Reset for the new key-value pair
            key = line.split(':')[0].strip(' "')
            value = []
        elif key is not None:
            # Accumulate values for the current key
            value.append(line)

    # Save the last key-value pair
    if key is not None:
        data[key] = '\n'.join(value).strip(' "')

    return data
````
